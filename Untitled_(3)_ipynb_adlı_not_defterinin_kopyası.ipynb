```python
import numpy as np
import random
import time
import sys
import os
import tensorflow as tf
import matplotlib.pyplot as plt
import matplotlib.pylab as pylab
import matplotlib.animation as animation
import mpl_toolkits.mplot3d.axes3d as p3
from os import listdir
from os.path import join
import numpy as np
import sys
from tensorflow.python.framework import ops
from google.colab import drive
drive.mount('/content/drive')
 
tf.compat.v1.disable_eager_execution()
 
 
joints_lines = [[20, 1, 2, 1, 8, 10, 2, 9, 11, 3, 4, 7, 7, 5, 6, 14, 15, 16, 17],
                [3, 3, 3, 8, 10, 12, 9, 11, 13, 4, 7, 5, 6, 14, 15, 16, 17, 18, 19]]
colorArray = np.random.rand(20,3,)
 
 
actionName = [
            "high_arm_wave",
            "horizontal_arm_wave",
            "hammer",
            "hand_catch",
            "forward_punch",
            "high_throw",
            "draw_x",
            "draw_tick",
            "draw_circle",
            "hand_clap",
            "two_hand_wave",
            "side_boxing",
            "bend",
            "forward_kick",
            "side_kick",
            "jogging",
            "tennis_swing",
            "tennis_serve",
            "golf_swing",
            "pick_up&throw"]
 
class ModelLSTM:
    def __init__(self, in_size, lstm_size, num_layers, out_size, session, learning_rate=0.003, name="rnn"):
        self.scope = name
        self.in_size = in_size
        self.lstm_size = lstm_size
        self.num_layers = num_layers
        self.out_size = out_size
 
        self.session = session
 
        self.learning_rate = tf.constant( learning_rate )
 
        # Last state of LSTM, used when running the network in TEST mode
        self.lstm_last_state = np.zeros((self.num_layers*2*self.lstm_size,))
 
        with tf.compat.v1.variable_scope(self.scope):
            self.xinput = tf.compat.v1.placeholder(tf.float32, shape=(None, None, self.in_size), name="inputLstm")
            self.lstm_init_value = tf.compat.v1.placeholder(tf.float32, shape=(None, self.num_layers*2*self.lstm_size), name="lstm_init_value")
 
            # LSTM
            self.lstm_cells = [ tf.compat.v1.nn.rnn_cell.BasicLSTMCell(self.lstm_size, forget_bias=1.0, state_is_tuple=False) for i in range(self.num_layers)]
            self.lstm = tf.compat.v1.nn.rnn_cell.MultiRNNCell(self.lstm_cells, state_is_tuple=False)
 
            # Iteratively compute output of recurrent network
            outputs, self.lstm_new_state = tf.compat.v1.nn.dynamic_rnn(self.lstm, self.xinput, initial_state=self.lstm_init_value, dtype=tf.float32)
 
            # Linear activation (FC layer on top of the LSTM net)
            self.rnn_out_W = tf.Variable(tf.random.normal( (self.lstm_size, self.out_size), stddev=0.01 ))
            self.rnn_out_B = tf.Variable(tf.random.normal( (self.out_size, ), stddev=0.01 ))
 
            outputs_reshaped = tf.reshape( outputs, [-1, self.lstm_size] )
            network_output = ( tf.matmul( outputs_reshaped, self.rnn_out_W ) + self.rnn_out_B )
 
            batch_time_shape = tf.shape(outputs)
            self.final_outputs = tf.reshape( network_output, (batch_time_shape[0], batch_time_shape[1], self.out_size) )
 
 
            ## Training: provide target outputs for supervised training.
            self.y_batch = tf.compat.v1.placeholder(
                tf.float32,
                (None, None, self.out_size)
            )
            y_batch_long = tf.reshape(self.y_batch, [-1, self.out_size])
            self.cost = tf.compat.v1.losses.mean_squared_error(y_batch_long, network_output)
            #self.cost = tf.reduce_mean(
            #    tf.nn.softmax_cross_entropy_with_logits(
            #        logits=network_output,
            #        labels=y_batch_long
            #    )
            #)
            self.train_op = tf.compat.v1.train.RMSPropOptimizer(self.learning_rate, 0.9).minimize(self.cost)
 
    ## Input: X is a single element, not a list!
    def run_step(self, x, init_zero_state):
        # Reset the initial state of the network.
        if init_zero_state:
            init_value = np.zeros((self.num_layers * 2 * self.lstm_size,))
        else:
            init_value = self.lstm_last_state
        out, next_lstm_state = self.session.run(
            [self.final_outputs, self.lstm_new_state],
            feed_dict={
                self.xinput: [x],
                self.lstm_init_value: [init_value]
            }
        )
        self.lstm_last_state = next_lstm_state[0]
        return out[0][0]
 
    def train_batch(self, xbatch, ybatch):
        init_value = np.zeros(
            (xbatch.shape[0], self.num_layers * 2 * self.lstm_size)
        )
        cost, _ = self.session.run([ self.cost, self.train_op], feed_dict={self.xinput:xbatch, self.y_batch:ybatch, self.lstm_init_value:init_value   } )
        return cost
    
    def check_restore_parameters(self, saver):
        ops.reset_default_graph()
        # Later, launch the model, use the saver to restore variables from disk, and
        # do some work with the model.
 
        saver.restore(self.session, 'drive/My Drive/MSRActionRecognition/model.ckpt.data-00000-of-00001')
 
 
class AnimateAction:
    def __init__(self,data,actionLabel):   
        self.data = data
        self.fig = plt.figure()
        self.figLabel = actionLabel
        self.fig.suptitle(('Action : ' + self.figLabel), fontsize=14, fontweight='bold')
        self.ax = p3.Axes3D(self.fig)
        self.x, self.y, self.z = self.create_bones(self.data)
        self.graph = self.ax.scatter([], [], [])
        self.reset_axes()
        
    def create_bones(self,data_arr):
        bones = []
        for j in range(data_arr.shape[0]):
            for i in range(19):
                bones.append([data_arr[j, joints_lines[0][i] - 1, 0:3], data_arr[j, joints_lines[1][i] - 1, 0:3]])
        x = np.zeros((data_arr.shape[0], 19, 2))
        y = np.zeros((data_arr.shape[0], 19, 2))
        z = np.zeros((data_arr.shape[0], 19, 2))
        for j in range(data_arr.shape[0]):
            i = 0
            for element in bones[19 * j: 19 * j + 19]:
                x[j][i] = [element[0][0], element[1][0]]
                y[j][i] = [element[0][2], element[1][2]]
                z[j][i] = [element[0][1], element[1][1]]
                i += 1
        return x, y, z
 
    def animate_skeleton(self,num):
        
        self.reset_axes()
        for index in range(19):
            self.ax.plot3D(self.x[num][index], self.y[num][index], self.z[num][index], label="graph", color=colorArray[index])
        self.graph._offsets3d = (self.data[num, :, 0], self.data[num, :, 2], self.data[num, :, 1])
 
    def reset_axes(self):
        self.ax.cla()
        self.ax.set_xlim3d([400.0, 0.0])
        self.ax.set_xlabel('X')
 
        self.ax.set_ylim3d([5000.0, 0.0])
        self.ax.set_ylabel('Z')
 
        self.ax.set_zlim3d([250.0, 0.0])
        self.ax.set_zlabel('Y')     
 
    def show_action(self):
        # Setting the axes properties      
        Writer = animation.writers['ffmpeg']
        writer = Writer(fps=10, metadata=dict(artist='Me'), bitrate=1800)
        ani = animation.FuncAnimation(self.fig, self.animate_skeleton, frames=self.data.shape[0], interval=10)
        ani.save('drive/My Drive/MSRActionRecognition/Motions/' + 'Motion_' + self.figLabel  +'.mp4', writer=writer) 
        plt.show()
 
 
class Data:
  
    def __init__(self):   
        self.path = 'drive/My Drive/MSRActionRecognition/dataset/MSRAction3D/MSRAction3DSkeleton(20joints)/'      
        self.allActions = np.zeros((567,20,3,), dtype=float)
 
    def name_parser(self, fname, sep_char):
        fnametostr = ''.join(fname).replace(self.path, '')
        ind = int(fnametostr.index(sep_char))
        label = int(fnametostr[ind + 1:ind + 3])
        return label
 
    def read_dataset(self):
        print('Loading MSR 3D Data, data directory %s' % self.path)
        data, labels, lens, subjects = [], [], [], []
        filenames = []
        documents = [join(self.path, d)
                     for d in sorted(listdir(self.path))]
        filenames.extend(documents)
        filenames = np.array(filenames)
 
        for file in filenames:
            #print("file",file)
            action = np.loadtxt(file)[:, :3].flatten()
            #print("action",action)
 
            labels.append(self.name_parser(file, 'a'))
            frame_size = len(action) // 60  # 20 iskeleton num x,y,z 3D points
            lens.append(frame_size)
            action = np.asarray(action).reshape(frame_size, 60)
            #print("action shape",action.shape)
 
            new_act = []
            for frame in action:
                new_act.append(frame)
 
            data.append(new_act)
            subjects.append(self.name_parser(file, 's'))
 
        data = np.asarray(data)
        labels = np.asarray(labels) -1
        lens = np.asarray(lens)
 
        subjects = np.asarray(subjects)
        print("All files read!")
        print('initial shapes [data label len]: %s %s %s' % (data.shape, labels.shape, lens.shape))
        return data,labels,lens,subjects
 
 
 
def main():
    #############################  DataSet  ##############################
    ops.reset_default_graph()
    ActdataSet = Data()
    actions = ActdataSet.read_dataset()
    actionsdata,labels,lens,subjects = actions
    
 
    #############################  InitNetwork  ##############################
    in_size = out_size = 20*3 # frame (x y z for 20 joints)
    lstm_size = 256 #128
    num_layers = 2
    batch_size = 1 #128
 
    NUM_TRAIN_BATCHES = 10000
 
    config = tf.compat.v1.ConfigProto()
    config.gpu_options.allow_growth=True
    sess = tf.compat.v1.InteractiveSession(config=config)
 
 
    net = ModelLSTM(in_size = in_size,
                        lstm_size = lstm_size,
                        num_layers = num_layers,
                        out_size = out_size,
                        session = sess,
                        learning_rate = 0.003,
                        name = "msr_rnn_network")
    
    sess.run(tf.compat.v1.global_variables_initializer())
    ckpt_file = "drive/My Drive/MSRActionRecognition/model2.ckpt"
    saver = tf.compat.v1.train.Saver(tf.compat.v1.global_variables())   
    #net.check_restore_parameters(saver)
 
    last_time = time.time()
 
    data_title_num = random.choice(range(len(actionsdata)))
    data = actionsdata[data_title_num]
    data =  np.array(data)
    data = data / 1000
    test_action = data
    #############################  Train  ##############################
    for i in range(NUM_TRAIN_BATCHES):
 
        time_steps = len(data)-1
        batch = np.zeros((1, time_steps, in_size))
        batch_y = np.zeros((1, time_steps, in_size))
        for j in range(time_steps):
            ind1 = j 
            ind2 = j+1
            batch[:, j, :] = data[ind1, :]        
            batch_y[:, j, :] = data[ind2, :]
        cst = net.train_batch(batch, batch_y)
        
        if (i%100) == 0:
            new_time = time.time()
            diff = new_time - last_time
            last_time = new_time
            print("batch: ",i,"   loss: ",cst)
            saver.save(sess, ckpt_file)
    print ("Loss: " , cst)
    #############################  Validation  ##############################
    predOutput = []
    out = net.run_step( [test_action[0]]  , True)    
    j = 1
    for j in range(len(test_action)):   
        out = net.run_step( [out]  , False)
        predOutput.append(out*1000)
 
    outAnim = np.array(predOutput).reshape(len(predOutput), 20,3)
    anim = AnimateAction(outAnim,actionName[labels[data_title_num]])
    anim.show_action()
    np.savetxt(('drive/My Drive/MSRActionRecognition/Motions/' + 'Motion_' + actionName[labels[data_title_num]] +'.txt'),np.array(predOutput), delimiter=' ')
 
 
if __name__ == "__main__":
    for i in range(10):
        main()
```
